```python
client = pymongo.MongoClient("mongodb://localhost:27017")
db = client["twitter"]
```

```python
db.tweets.count_documents({})
```

```python
col1 = db.tweets.find(
  {
    'retweet_count' : {'$gt' : 100000}
  }
)

for i in col1:
    print(i)
```
